```python
def to_int(valor):
    try:
        return int(valor.replace(',','.'))
    except:
        valor
```

```python
from datetime import datetime

def to_date(valor):
    try:
        fecha = datetime.strptime(valor, '%Y-%m-%d').date()
        return fecha
    except:
        return valor
```
